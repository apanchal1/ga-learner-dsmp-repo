```python
#  create dataframe where Weather column contains fog in it.
weather[(weather['Weather'].str.contains('fog'))]
```

```python
#  create dataframe where Weather column does-not-contains 'fog' in it.
weather[(~ weather['Weather'].str.contains('fog'))]
```

```python
#  creating wind_log column which is log of Wind speed
weather['wind_log'] = np.log(weather['Wind Spd (km/h)'])
```

```python
#  drop wind_log column
weather = weather.drop(['wind_log'],axis=1)
```

```python
#  create new column Capital_Weather which is 'Weather' column upper case
weather['Capital_Weather'] = weather['Weather'].str.upper()
```

```python
weather.shape
```

```python
#  drop Capital_Weather column
weather = weather.drop(['Capital_Weather'],axis=1)
```

```python
#  convert Temp column from Celcius to Freheneits
weather['Temp (C)']= (weather['Temp (C)']*9/5)+32
```

```python
#  renameing columns in dataframe
weather.rename(columns={'Temp (C)':'Temp (F)'},inplace=True)
```

```python
weather
```

```python
#  sorting 'dew point Temp(C)' column
weather['Dew Point Temp (C)'].sort_values(ascending=False)
```

```python
#  dataframe contains top ten highest temperature and assign to top_10
top_10 = weather.loc[weather['Temp (F)'].sort_values(ascending=False).index].head(10)
```

```python
# take out unique values for weather columns in top_10 dataframe
top_10['Weather'].unique()
```

```python
df1
```

```python
df2
```

```python
df3 = pd.concat((df1,df2))
df3
```
