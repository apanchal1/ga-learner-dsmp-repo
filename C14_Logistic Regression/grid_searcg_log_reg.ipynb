```python
import pandas as pd
```

```python
df = pd.read_csv("breast_cancer.csv")
```

```python
df.head()
```

```python
df.drop(["Unnamed: 32", "id"],1,inplace = True)
```

```python
df.head()
```

```python
df.shape
```

```python
train_test = df[:500]
validation = df[500:]
```

```python
train_test.shape
```

```python
validation.shape
```

```python
y = train_test["diagnosis"]
X = train_test.drop(["diagnosis"],1)
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
scaler = StandardScaler()
le = LabelEncoder()
```

```python
y = le.fit_transform(train_test["diagnosis"])
X = pd.DataFrame(scaler.fit_transform(X), columns=list(X))
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split as tts, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
knn = KNeighborsClassifier()
```

```python
import numpy as np
params = {"n_neighbors":np.arange(2,15,1), 
         "metric":["euclidean", "minkowski", "manhattan", "jaccard", "cosine"]}
```

```python
params
```

```python
knn_cv = GridSearchCV(knn, param_grid=params, cv = 10)
```

```python
knn_cv.fit(X,y)
```

```python
knn_cv.best_params_
```

```python
knn_cv.best_score_
```

```python
best_knn_model = knn_cv.best_estimator_
```

```python
validation
```

```python
y_validation= validation["diagnosis"]
X_validation = validation.drop(["diagnosis"],1)
y_validation = le.fit_transform(validation["diagnosis"])
X_validation = pd.DataFrame(scaler.fit_transform(X_validation), columns=list(X_validation))
```

```python
X_validation.shape
```

```python
y_validation.shape
```

```python
y_pred = best_knn_model.predict(X_validation)
```

```python
y_validation
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
```

```python
print(classification_report(y_validation,y_pred))
```

```python
print(roc_auc_score(y_validation,y_pred))
```

```python
print(accuracy_score(y_validation,y_pred))
```

```python
print (confusion_matrix(y_validation,y_pred))
```
