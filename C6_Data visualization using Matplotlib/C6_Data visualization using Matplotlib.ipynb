Since we want to compare amongst months, we need one value for each month.

Generate a line chart visualize the temperature readings in the month of january

x-values should be the values should be the months, y-values should be mean values in the Temp(C) column per month,
Display the plot

```python
weather['Date/Time'] = pd.to_datetime(weather['Date/Time'])
```

```python
weather.dtypes
```

```python
import datetime
```

```python
temp = []
for x in range(1,13):
    temp.append(weather[weather['Date/Time'].dt.month==x]['Temp (C)'].mean())
```

```python
plt.plot(temp,marker = '*')
plt.xlabel('Month')
plt.ylabel('Temp')
plt.show()
```

# Working with auto.csv dataset

```python
auto = pd.read_csv('Automobile_data.csv')
```

```python
auto.head()
```

```python
auto.dtypes
```

```python
#  creating function to convert symbolic data into mean of that columns
def missing_values(column,symbol):
    mean_value = auto[auto[column]!=symbol][column].astype(float).mean()
    auto[column] = auto[column].str.replace(symbol,str(mean_value))
    return auto
```

```python
missing_values('price','?')
```

```python
auto['price'].value_counts()
```
