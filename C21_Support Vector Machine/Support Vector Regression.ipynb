```python
numerical_df = auto.select_dtypes(include='number')
```

```python
categorical_df = auto.select_dtypes(exclude='number')
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean')
numerical_df = pd.DataFrame(imputer.fit_transform(numerical_df),columns = numerical_df.columns)
```

```python
imputer = SimpleImputer(strategy='most_frequent')
categorical_df = pd.DataFrame(imputer.fit_transform(categorical_df),columns = categorical_df.columns)
```

```python
categorical_df
```

```python
auto = pd.concat([numerical_df,categorical_df],axis=1)
```

```python
auto.isnull().sum()
```

```python
auto.dtypes
```

```python
auto = pd.get_dummies(auto)
```

```python
auto
```

```python
col = auto.columns
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
auto = pd.DataFrame(sc_X.fit_transform(auto),columns= col)
```

```python
auto
```

```python
X = auto.drop('price',1)
y = auto['price']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test, y_train, y_test = tts(X,y,test_size=0.2,random_state = 42)
```

```python
from sklearn.svm import SVR
```

```python
regressor = SVR(kernel='linear')
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_pred,y_test)
```
