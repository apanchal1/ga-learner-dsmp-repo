#  Function of  finding Binomial Function given n , r and p
b(x; n, P) = nCx * P^x * (1 â€“ P)^n â€“ x

```python
import math
def binomial_f(n,r,p):
    q = 1-p
    numerator = math.factorial(n)*(p**r)*(q**(n-r))
    denominator = math.factorial(r)*math.factorial(n-r)
    return numerator/denominator
```

```python
binomial_f(11,7,0.38)
```

#  KNN using Pandas for diabetes dataset 

```python
diabetes = pd.read_csv('diabetes.csv')
diabetes.head()
```

```python
training_set = diabetes.iloc[:500,:] 
```

```python
testing_set = diabetes.iloc[501:,:]
```

```python
from scipy.spatial.distance import euclidean
```

```python
def euclidean_distance(training_set,testing_set):
    l =[]
    for i in range(0,training_set.shape[0]):
        l.append(euclidean(training_set.iloc[i],testing_set))
    l  = pd.Series(l)
    return l
```

```python
result = euclidean_distance(training_set,testing_set.iloc[0])
training_data_idx = list(result.sort_values().iloc[:5].index)
```

```python
count0=0
count1=0
for i in training_data_idx:
    if(training_set["diabetes"].iloc[i]==1):
        count1=count1+1
    else:
        count0=count0+1
if(count1>count0):
    print("1")
else:
    print("0")
```
