```python
# Load the ipl_mathces_small.csv data 
import csv
data_2 =[]
with open(r'C:\Users\Ajay\Desktop\ipl_matches_small.csv','r') as f:
    data = csv.reader(f)
    for row in data:
        data_2.append(row)
data_2
```

```python
# take out unique match_code from data located at column 0
match_codes = list(set([x[0] for x in data_2[1:]]))
match_codes
```

```python
# take out total no. teams playing in the tournaments

team_names_1 = [x[3] for x in data_2[1:]]
print(set(team_names_1))

team_names_2 = [x[4] for x in data_2[1:]]
print(set(team_names_2))

team_names = team_names_1+team_names_2
print(list(set(team_names)))
```

```python
#  run scored by batsman and strike rate (runs scored by player/ ball faced)
runs_by_batsman = sum([int(x[16]) for x in data_2[1:] if x[13]=='ST Jayasuriya'])
strike_rate = runs_by_batsman/len([int(x[16]) for x in data_2[1:] if x[13]=='ST Jayasuriya'])*100
strike_rate
```

```python
#  define function to calculate strike rate for batsman
def batsman_strike_rate(batsman_name):
    try:
        runs_by_batsman = sum([int(x[16]) for x in data_2[1:] if x[13]==batsman_name])
        strike_rate = runs_by_batsman/len([int(x[16]) for x in data_2[1:] if x[13]==batsman_name])*100
        return strike_rate
    except:
        return ('Batsman not in list')

print(batsman_strike_rate('ST Jayasuriya'))
```
