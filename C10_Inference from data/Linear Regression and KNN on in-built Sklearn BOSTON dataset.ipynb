```python
from sklearn.datasets import load_boston
import pandas as pd
```

```python
data = load_boston()
```

```python
cols = list(data.feature_names)
df = pd.DataFrame(data.data)
```

```python
## To remove warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
df.columns = cols
df['MEDV'] = data.target
```

```python
df
```

```python
print(data.DESCR)
```

```python
cols_list = list(df)
```

```python
cols_list
```

```python
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsRegressor
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

## Normalizing Dataset using MinMaxScalar

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
df = scalar.fit_transform(df)
```

```python
df = pd.DataFrame(df)
df.columns = cols_list
```

```python
df
```

## Linear Regressing to find R-square

```python
lin_reg = LinearRegression()
knn = KNeighborsRegressor()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(),annot=True,cmap = 'YlGnBu')
```

```python
x = df[['RM','PTRATIO',"LSTAT"]]
y = df['MEDV']
```

```python
x_train, x_test, y_train, y_test = tts(x,y,test_size=0.3,random_state=42)
```

```python
lin_reg.fit(x_train,y_train)
```

```python
y_pred = lin_reg.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
## Find RMSE
from sklearn.metrics import mean_squared_error
from math import sqrt

rms = sqrt(mean_squared_error(y_test, y_pred))
rms
```

## OLS Method to find R-Squared and P-value

```python
df_ols = sm.add_constant(df)
```

```python
df_ols.head()
```

```python
X = df_ols.drop(['MEDV'],1)
model = sm.OLS(y,X).fit()
model.summary()
```

## Using RFE (Recursive Feature Elimination)

```python
from sklearn.feature_selection import RFE
```

```python
df
```

```python
x_rfe = df.drop(['MEDV'],1)
y_rfe = df['MEDV']
lin_reg = LinearRegression()
```

```python
rfe = RFE(lin_reg,4)
fit = rfe.fit(x_rfe,y_rfe)
print(f'Number of features: {fit.n_features_}')
print(f'Selected features: {fit.support_}')
print(f'Ranking the features: {fit.ranking_}')
```
